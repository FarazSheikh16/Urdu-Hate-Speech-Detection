```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, cohen_kappa_score

# Replace 'annotator1_csv_file' and 'annotator2_csv_file' with your actual file paths
annotator1_csv_file = '/content/Faraz Labels.xlsx'
annotator2_csv_file = '/content/Jibran Labels.xlsx'

# Read labels and text from CSV files
df_annotator1 = pd.read_excel(annotator1_csv_file)
df_annotator2 = pd.read_excel(annotator2_csv_file)

# Assuming 'text' is the column containing text data, and 'Label (h/o)' is the label column
annotator1_labels = df_annotator1['Label (h/o)'].tolist()
annotator2_labels = df_annotator2['Label (h/o)'].tolist()
texts = df_annotator1['Text']
print(texts[1])

# Calculate confusion matrix
conf_matrix = confusion_matrix(annotator1_labels, annotator2_labels, labels=['h', 'o', 'n'])

# Calculate Cohen's Kappa
kappa = cohen_kappa_score(annotator1_labels, annotator2_labels)

# Identify discrepancies and store in a DataFrame
discrepancies = pd.concat([
    pd.DataFrame({'Text': texts[i], 'label_1': annotator1_labels[i], 'label_2': annotator2_labels[i]}, index=[i])
    for i in range(len(texts)) if annotator1_labels[i] != annotator2_labels[i]
], ignore_index=True)

# Display the confusion matrix with labels
class_labels = ['h', 'o', 'n']
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix,
                               display_labels=class_labels)
disp.plot()
plt.show()

# Display discrepancies DataFrame
print("\nDiscrepancies:")
print(discrepancies)

print(f"\nCohen's Kappa: {kappa}")
```
